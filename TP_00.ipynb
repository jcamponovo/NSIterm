# <span style="color:red;"><center>Le code ISBN d'un livre </center></span>

## 1. Le standard ISBN
L'ISBN (acronyme d'International Standard Book Number) est un numéro international normalisé permettant **d'identifier de façon unique un livre**, quel que soit son support de publication,imprimé ou numérique. Même deux livres avec un titre identique ont des codes ISBN ditincts.

**L'ISBN permet de simplifier le traitement informatisé des livres.** Grâce à lui, les libraires peuvent passer des commandes standardisées, les distributeurs ont le même code pour traiter les commandes et les retours, les différentes opérations de gestion dans les bibliothèques et centres de documentation sont également facilitées.

Par ailleurs, le caractère international de cette numérotation constitue, à l'étranger également, une référence unique pour tous les professionnels du livre.

L’ISBN est un numéro à 13 chiffres depuis le 1er janvier 2007 (les ouvrages antérieurs avaient un ISBN à 10 chiffres).

![Code ISBN](img/Code_ISBN.png)

## 2. Signification d'un code ISBN
Prenons le code ISBN à **13 chiffres** de l'exemple précédent : 978-2-1234-5680-3
* Les **3 premiers chiffres** 978, signifie que l’article est un **livre** (979 peut aussi être employé en cas de besoin). Dans un code barre, ils indiquent le pays producteur mais comme 978 et 979 ne sont pas utilisés, on les a attribués aux livres pour le pays "Bookland".
*  Le **chiffre suivant** est le numéro identifiant le **groupe national, linguistique ou géographique**. Ici, le 2 identifie la francophonie dans son ensemble.
* Le **nombre suivant**, ici 1234, identifie un **éditeur** en particulier.
* Vient ensuite le **numéro d'identification** de cette édition en particulier, ici 5680.
* La **dernière partie** de l’ISBN, 3 ici, est la **clé de contrôle**. Elle est reliée aux précédents numéros du code ISBN par un algorithme mathématique qui permet d’en vérifier la validité. Elle peut varier de 0 à 9.

## 4. Comment est calculée la clé de contrôle ?
* *1ère étape :* Un coefficient (toujours le même) est attribué à chacun des 12 premiers chiffres du code ISBN de façon à faire la $somme$ pondérée de ces 13 premiers chiffres.

Exemple avec le code 978-2-1234-5680-3 :

|Chiffre ISBN  |9|7 |8|2|1|2|3|4 |5|6 |8|0|
|    -         |-|- |-|-|-|-|-|- |-|- |-|-|
|Coefficient   |1|3 |1|3|1|3|1|3 |1|3 |1|3|
|Multiplication|9|21|8|6|1|6|3|12|5|18|8|0|

$Somme = 9+21+8+6+1+6+3+12+5+18+8+0 = 97$


* *2ème étape :* On calcule le $reste$ de la division entière par 10 de la $somme$. Si ce $reste$ est nul alors la $clé$ de contrôle est 0, sinon la $clé$ vaut $10 - reste$.

Exemple avec le code 978-2-1234-5680-3 :


``` python
97//10 = 9
97%10 = 7
10 - 97%10 = 3
```
**La clé est donc bien égale à 3.**

# 5. Description de l'algorithme

Il nous faut une fonction telle que :
* **Données en entrée** : Code ISBN de 13 chiffres
* **Variables** à utiliser : somme, reste, cle
* **Résultat** : vrai si le code est valide, faux sinon
* **Instructions** :
    * lire le code ISBN
    * calculer la somme pondérée
    * si somme%10 == 0 alors cle = 0
    * sinon cle = 10 - somme%10
    * si cle == dernier chiffre du code ISBN alors retourner vrai
    * sinon retourner faux
